```python
import re
import math
from shapely.geometry import LineString

f = open('input.txt', 'r')

directions = f.read().strip().split(', ')

print(directions)
```

```python
card = []
dist = []

for direction in directions:
    card.append(direction[0])
    dist.append(int(re.findall("\d+", direction)[0]))

path = zip(card, dist)

coord = [0, 0]
hist = []

state = 'N'

for action, dist in path:
    if(tuple(coord) in hist):
        print('Duplicate found', coord)
    hist.append(tuple(coord))


    if((state is 'S' and action is 'L') or (state is 'N' and action is 'R')):
        # if moving east add east
        coord[0] += dist
        state = 'E'
    elif((state is 'W' and action is 'L') or (state is 'E' and action is 'R')):
        # if moving south
        coord[1] -= dist
        state = 'S'
    elif((state is 'S' and action is 'R') or (state is 'N' and action is 'L')):
        # if moving west
        coord[0] -= dist
        state = 'W'
    elif((state is 'W' and action is 'R') or (state is 'E' and action is 'L')):
        # if moving north
        coord[1] += dist
        state = 'N'
    else:
        raise Exception('error in state and action')
    
        
print(coord)
abs(coord[0] + abs(coord[1]))
print('\n', hist)
```

```python
import numpy as np
import pandas as pd
f = open('input.txt', 'r')

n = f.read().strip().split('\n') # this could maybe be optimized, but I don't care enough

triList = []
```

```python
def check():
    for line in n:
        sides = line.split()
        sides = [int(i) for i in sides]
        triList.append(sides)
#         triList.append(checkTriangle(sides))
        
```

```python
def checkTriangle(sideLengthList):
    counter = 0
    for side in sideLengthList:
        if((sum(sideLengthList) - side) > side):
            counter += 1
        else:
            return(False)
        if(counter == 3):
            return(True)
```

```python
checkTriangle([3, 4, 1])
```

```python
check()
```

```python
small = triList
small = np.array(small)
small = pd.DataFrame(small)
#small.columns = ['a', 'b', 'c']
small
```
